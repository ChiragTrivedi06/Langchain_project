```python
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
import evaluate

# 1. Load dataset (IMDb: binary sentiment classification)
dataset = load_dataset("imdb")

# 2. Tokenizer
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True, padding="max_length", max_length=128)

tokenized_dataset = dataset.map(tokenize_function, batched=True)

# 3. Model
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)

# 4. Metrics
accuracy = evaluate.load("accuracy")
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = logits.argmax(axis=-1)
    return accuracy.compute(predictions=predictions, references=labels)

# 5. Training setup
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=2,
    weight_decay=0.01,
    logging_dir="./logs",
    push_to_hub=False
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"].shuffle(seed=42).select(range(5000)),  # small subset
    eval_dataset=tokenized_dataset["test"].shuffle(seed=42).select(range(1000)),
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

trainer.train()
```

```python
# BERT (Contextual Word Embeddings)
from transformers import AutoTokenizer, AutoModel
import torch

# Load BERT
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

# Example sentence
text = "I went to the bank to deposit money."

# Tokenize
inputs = tokenizer(text, return_tensors="pt")

# Get embeddings
with torch.no_grad():
    outputs = model(**inputs)
    embeddings = outputs.last_hidden_state  # shape: (batch, seq_len, hidden_dim)
print(inputs)
print("Shape of embeddings:", embeddings.shape)
print("Embedding for 'bank':", embeddings[0, 4, :10])
```

```python
# sentense embedding we can compare3 similarity between sentense
!pip install sentence-transformers -q
from sentence_transformers import SentenceTransformer, util

# Load pre-trained model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Example sentences
sent1 = "I love playing football."
sent2 = "Soccer is my favorite sport."
sent3 = "The sky is blue today."
sent4 = "your wear blue t-shirt today and Sky is also blue."

# Get embeddings
embeddings = model.encode([sent1, sent2, sent3,sent4])

# Compare similarity
similarity = util.cos_sim(embeddings[0], embeddings[1])
print("Similarity between sent1 & sent2:", similarity.item())

similarity2 = util.cos_sim(embeddings[0], embeddings[2])
print("Similarity between sent1 & sent3:", similarity2.item())
```

```python
from langchain.chat_models import init_chat_model
from langchain.agents import load_tools, initialize_agent

# Gemini LLM setup (with proper authentication)
llm = init_chat_model("gemini-2.5-flash", model_provider="google_genai")

# Define your supplier database
supplier_db = [
    {"name":"GreenFabrics Ltd", "product":"Organic Cotton Fabric", "rating":4.9, "country":"India", "verified":True},
    {"name":"AceTextiles Inc", "product":"Recycled Polyester Yarn", "rating":4.7, "country":"China", "verified":True},
    # More entries as needed
]

def find_supplier(product_type):
    return [
        s for s in supplier_db
        if product_type.lower() in s["product"].lower() and s["verified"]
    ]

# Load your custom tool for supplier search
tools = load_tools(["supplier_search"], db=supplier_db)

# Initialize the LangChain agent (with Gemini)
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent_type="zero-shot-react"
)

question = "Find trustworthy suppliers for organic cotton fabric in India"
answer = agent.run(question)
print(answer)

```
