```python
from langchain.tools import tool
from langchain_community.tools.tavily_search import TavilySearchResults
@tool("vector_search", return_direct=False)
def vector_search_tool(query: str) -> str:
    ''' call your retriever here'''
    results = retriever.invoke(query)
    return "\n".join([doc.page_content[:300] for doc in results])

def wiki_tool(query: str) -> str:
    import wikipedia
    return wikipedia.summary(query, sentences=2)

def tavily_tool(query: str) ->list:

    tool = TavilySearchResults()
    return tool.invoke({"query": f"{query}"})
```

```python
from langchain import hub
tools = [vector_search_tool]
# 3. Use LangChain’s built-in ReAct prompt template
prompt = hub.pull("hwchase17/react")

tools = [
    Tool(
        name="vector_search",
        func=vector_search_tool,
        description="Use this to search the Virat Kohli dataset for factual answers."
    ),
    Tool(
        name="wiki_search",
        func=wiki_tool,
        description="Use this to get general info from Wikipedia."
    ),
    Tool(
        name = "add",
        func = add,
        description = "when i pass addition operator."
    ),
    Tool(
        name = "tavily_search",
        func=tavily_tool,
        description="search on web"
    )
]
```

```python
from langchain_core.output_parsers import JsonOutputParser
from pydantic import BaseModel, Field

# Define strict schema with Pydantic
class AnswerSchema(BaseModel):
    answer: str = Field(description="The final answer from the dataset or tools")
    source: str = Field(description="Where the answer came from (vector_search, wiki_search, etc.)")

# Parser based on schema
parser = JsonOutputParser(pydantic_object=AnswerSchema)
```

```python
from langchain.agents  import create_react_agent, AgentExecutor
agent_prompt = PromptTemplate.from_template(
     """
You are an assistant that answers ONLY from the Virat Kohli dataset.
Rules:
1. ALWAYS start with 'vector_search' tool.
2. NEVER use outside knowledge. If no info, say: "I don’t have knowledge about this document."
3. Cite source and page number like [Source | Page N].
4. Be detailed but ONLY within dataset.
You have access to the following tools:
{tools}

Use them carefully. Tool names you can call: {tool_names}.
Question: {input}

{agent_scratchpad}
"""
 )
agent_prompt = agent_prompt.partial(format_instructions=parser.get_format_instructions())

react_agent = create_react_agent(llm=llm, tools=tools, prompt=agent_prompt)
print(react_agent)
executor = AgentExecutor(agent = react_agent, tools = tools,verbose=True, handle_parsing_errors=True)
print(executor)
```

```python
react_agent = create_react_agent(llm=llm, tools=tools, prompt=prompt)

# 5. Executor with safety limits
executor = AgentExecutor(
    agent=react_agent,
    tools=tools,
    verbose=True,
    handle_parsing_errors=True,
    max_iterations=5
)
```

```python
# 6. Run query
response = executor.invoke({"input": "who is prime minister of india?"})
print(response)

```
